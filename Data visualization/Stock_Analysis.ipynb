```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data,wb
import datetime
import seaborn as sns
%matplotlib inline
```

```python
start=datetime.datetime(2010,1,1)
end=datetime.datetime(2020,1,1)
```

# get the data for the following


GOOGLE (GOOGL),
AMAZON (AMZN),
MICROSOFT(MSFT),
APPLE (AAPL)


```python
google=data.get_data_yahoo('GOOGL',start,end)
google.head()
```

```python
amazon=data.get_data_yahoo('AMZN',start,end)
amazon.head()
```

```python
microsoft=data.get_data_yahoo('MSFT',start,end)
microsoft.head()
```

```python
apple=data.get_data_yahoo('AAPL',start,end)
apple.head()
```

```python
tickers=['GOOGL','AMZN','MSFT','AAPL']
df=pd.concat([google,amazon,microsoft,apple],axis=1,keys=tickers)
df.head()
```

```python
df.columns.names=['Ticker Name','Stock Info']
df.head()
```

** What is the max Close price for each companystock throughout the time period?**

```python
df.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each company
stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns=pd.DataFrame()
for val in tickers:
    returns[val+' Returns']=df[val]['Close'].pct_change()
returns.head()
```

***Create a pair plot for the returns data frame***

```python
sns.pairplot(returns[1:])
```

```python
returns.idxmax()
```

```python
returns.idxmin()
```

```python
returns.std()
```

*** Find the standard deviation for year 2011 returns ***

```python
returns.loc['2011-01-01':'2011-12-31'].std() #observation : As the std for amazon is more, there is fall of stock in that particular year
```

****Create a distribution plot for the 2011 returns of AMAZON****

```python
sns.distplot(returns['AMZN Returns'].loc['2011-01-01':'2011-12-31'])
```
